# 1. Objective

To fit mixed-effects models for data nested within clusters. The data contains a certain number of clusters within which a random sample of observations has been collected.

Covariates can measure the nature of clusters or of the individual data

The steps involved in fitting a mixed-effects model include:

1. Fit a maximal fixed-effects structure
2. Include random effects and check if random effects are significant
3. Once random effects structure is fixed check if all fixed-effects are needed

Usually, we are looking to disprove hypotheses around fixed effects, by including the nested variance structure, we make our estimates more reliable.

# 2. Two-level nested data

```r
data_df = read.delim("data/rat_pup.dat.txt")
```

```r
glimpse(data_df)
```

In this experiment, each litter is randomly assigned to Control or Treatment and the units of analysis - pups are ensconsed within each litter.

```r
data_df %>%
    group_by(treatment, sex) %>%
    summarize(num_obs = n(),
              mean_weight = mean(weight),
              sd_weight = sd(weight),
              min_weight = min(weight),
              max_weight = max(weight))
```

```r
data_df %>%
    mutate(female = ifelse(sex == "Female", 1, 0)) ->
    data_df
```

Formally:

For pup $i$ in litter $j$, 

$$weight_{ij} = \beta_0 + \beta_1 treatment_{1j} + \beta_2 treatment_{2j} + \beta_3 female_{ij} + \beta_4 litsize_j + \\
                \beta_5 treatment_{1j} \times female_{ij} + \beta_6 treatment_{2j} \times female_{ij} + \\
                u_j + \epsilon_{ij}$$
                
$$ u_j \sim N(0, \sigma_{litter}^2) , \epsilon_{ij} \sim N(0, \sigma_{treatment}^2)$$

Fit a maximal fixed-effects structure

```r
base_model = glmmTMB(weight ~ treatment * female + litsize + 
                              (1|litter),
                     family = "gaussian",
                     data = data_df,
                     REML = TRUE)
```

```r
plot(simulateResiduals(base_model))
```

Check if random effects are necessary

```r
no_ranef_model = glmmTMB(weight ~ treatment * female + litsize,
                         data = data_df,
                         family = "gaussian",
                         REML = TRUE)
```

```r
anova(base_model, no_ranef_model)
```

So, we need to include random effects

We now add a variance structure

```r
base_model = lme(weight ~ treatment * female + litsize,
                 random = ~ 1 | litter,
                 data = data_df,
                 method = "REML")
```

```r
diff_var_mixed_model = lme(weight ~ treatment * female + litsize,
                           random = ~ 1 | litter,
                           weights = varIdent(form = ~ 1 | treatment),
                           data = data_df,
                           method = "REML")
```

```r
anova(base_model, diff_var_mixed_model)
```

```r
summary(diff_var_mixed_model)
```

Reduce fixed-effects

```r
reduced_fixef_model = lme(weight ~ treatment + female + litsize,
                          random = ~ 1 | litter,
                          weights = varIdent(form = ~ 1 | treatment),
                          data = data_df)
```

```r
summary(reduced_fixef_model)
```

```r
intervals(reduced_fixef_model)
```

```r
getVarCov(reduced_fixef_model, individual = "3", type = "marginal")
```

# 3. Three-level nested data

```r
data_df = read_csv("data/classroom.csv")
```

```r
glimpse(data_df)
```

Here, the unit of analysis are students who are nested within classes that are inturn nested within schools. Each of these levels has their own covariates.

Formally:

For student $i$ in class $j$ within school $k$:

$$mathgain_{ijk} = \beta_0 + \beta_1 mathkind_{ijk} + \beta_2 sex_{ijk} + \beta_3 minority_{ijk} + \beta_4 ses_{ijk} + \\
                    \beta_5 yearstea_{jk} + \beta_6 mathprep_{jk} + \beta_7 mathknow_{jk} + \\
                     u_k + u_{j|k} + \epsilon_{ijk}$$
                     
$$u_k \sim N(0, \sigma_{int:school}^2), u_{j|k} \sim N(0, \sigma_{int:class}^2), \epsilon_{ijk} ~ N(0, \sigma^2)$$

```r
base_model = lme(mathgain ~ mathkind + sex + minority + ses + yearstea + mathprep + mathknow + housepov,
                 random = ~ 1 | schoolid/classid,
                 data = data_df,
                 na.action = na.omit,
                 method = "REML")
```

```r
summary(base_model)
```

```r
proposed_model = lme(mathgain ~ mathkind + sex + minority + ses,
                     random = ~ 1 | schoolid/classid,
                     data = data_df,
                     na.action = na.omit,
                     method = "REML")
```

```r
plot(proposed_model, col = 1)
```

```r
summary(proposed_model)
```

```r
intervals(proposed_model)
```
