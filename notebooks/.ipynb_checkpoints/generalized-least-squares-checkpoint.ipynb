# 1. Objective

Dealing with heterogeneity in models by using generalized least squares. We are looking to address varying residual patterns in various strata of data. So, the assumption of homogenity of variance at the sample level is violated.

# 2. Variance specification involving one factor or numeric variable

We begin with an introductory example

```r
data_df = read.delim("data/Squid.txt") %>% mutate(fMONTH = factor(MONTH)) %>% drop_na()
```

```r
glimpse(data_df)
```

```r
lm_model1 = lm(Testisweight ~ DML * fMONTH,
               data = data_df)
```

```r
plot(fitted(lm_model1), residuals(lm_model1))
```

When we see such plots, we should never make any inference from the model coefficients. Though the homogeneity is violated, the variance shows a clear pattern - it increases with increasing values of the outcome. 

Since number of predictors is less, we can see how the residual pattern is across different predictors.

```r
plot(data_df$fMONTH, residuals(lm_model1), xlab="Month", ylab="Residuals")
```

```r
plot(data_df$DML, residuals(lm_model1), xlab="DML", ylab="Residuals")
```

We recover the original pattern observed in residuals, and it seems to increase with increasing DML. We can incorporate this into the model.

```r
?varClasses
```

## 2.1. Variance as a function of a numeric variable

Here we instruct the model to measure the changes in variance by the values of numeric variable using `varFixed`

```r
gls_lm1 = gls(Testisweight ~ DML * fMONTH,
              data = data_df)

gls_fixedvar1 = gls(Testisweight ~ DML * fMONTH,
                    weights = varFixed(~ DML),
                    data = data_df)
```

```r
anova(gls_lm1, gls_fixedvar1)
```

Notice that the models are not nested so we do not have F values

```r
AIC(gls_lm1, gls_fixedvar1)
```

```r
plot(gls_fixedvar1, col = 1)
```

Does not seem correct for heterogeneity

## 2.2. Variance as a function of a factor

We account for difference in spread across different strata as defined by the categorical variables

```r
gls_factorvar1 = gls(Testisweight ~ DML * fMONTH, 
                     weights = varIdent(form = ~1 | fMONTH),
                     data = data_df)
```

```r
anova(gls_lm1, gls_fixedvar1, gls_factorvar1)
```

```r
plot(gls_factorvar1, col = 1)
```

The issue of heterogeneity still remains, we will need a stronger variance specification

## 2.3. Non-linear variance specification for a numeric variable

### 2.3.1. Variance covariate cannot take zero values

```r
gls_powervar1 = gls(Testisweight ~ DML *fMONTH,
                    weights = varPower(form = ~ DML),
                    data = data_df)
```

```r
gls_powervar2 = gls(Testisweight ~ DML * fMONTH,
                    weights = varPower(form = ~ DML|fMONTH),
                    data = data_df)
```

```r
anova(gls_fixedvar1, 
      gls_factorvar1,
      gls_powervar1, 
      gls_powervar2)
```

```r
plot(gls_powervar2, col = 1)
```

This seems to rectify things a bit

```r
summary(gls_powervar2)
```

### 2.3.2. Variance covariate can take zero values

```r
gls_powervar3 = gls(Testisweight ~ DML * fMONTH,
                    weights = varExp(form = ~ DML),
                    data = data_df)
```

```r
gls_powervar4 = gls(Testisweight ~ DML * fMONTH,
                    weights = varExp(form = ~ DML|fMONTH),
                    data = data_df)
```

```r
plot(gls_powervar4, col = 1)
```

## 2.4. Mixed specification for variance structures

```r
gls_mixed1 = gls(Testisweight ~ DML * fMONTH,
                 weights = varConstPower(form = ~DML),
                 data = data_df)
```

```r
gls_mixed2 = gls(Testisweight ~ DML * fMONTH,
                 weights = varConstPower(form = ~ DML|fMONTH),
                 data = data_df)
```

```r
plot(gls_mixed2, col = 1)
```

## 2.5. Final model diagnostics

```r
plot(gls_powervar2, col = 1)
```

```r
coplot(residuals(gls_powervar2, type = "normalized") ~ DML|fMONTH,
       data = data_df,
       ylab = "Normalized residuals")
```

Convince yourself that the residual pattern is now gone

```r
summary(gls_powervar2)
```

# 3. Variance specification with multiple factor or numeric variables

```r
data_df = read.delim("data/Biodiversity.txt") %>% mutate(fTreatment = factor(Treatment), fNutrient = factor(Nutrient))
```

```r
glimpse(data_df)
```

```r
model_formula = formula(Concentration ~ Biomass * fTreatment * fNutrient)
```

```r
gls_lm1 = gls(model_formula, data = data_df)
```

```r
gls_factorvars1 = gls(model_formula,
                      weights = varIdent(form = ~ 1 | fTreatment * fNutrient),
                      data = data_df)
```

```r
gls_factorvars2 = gls(model_formula,
                      weights = varIdent(form = ~ 1 | fNutrient),
                      data = data_df)
```

```r
gls_factorvars3 = gls(model_formula,
                      weights = varIdent(form = ~ 1 | fTreatment),
                      data = data_df)
```

```r
anova(gls_lm1,
      gls_factorvars1,
      gls_factorvars2,
      gls_factorvars3)
```

```r
plot(gls_factorvars1, col = 1)
```

```r
summary(gls_factorvars1)
```

```r
anova(gls_factorvars1)
```

# 4. Model selection

```r
var_structure = varIdent(form = ~ 1 | fTreatment * fNutrient)
```

```r
full_model = gls(Concentration ~ Biomass + fTreatment + fNutrient + 
                                 Biomass:fTreatment + 
                                 Biomass:fNutrient + 
                                 fTreatment:fNutrient,
                 weights = var_structure,
                 method = "ML",
                 data = data_df)
```

Start dropping interactions one by one

```r
partial_model1 = gls(Concentration ~ Biomass + fTreatment + fNutrient + 
                                     Biomass:fNutrient + 
                                     fTreatment:fNutrient,
                     weights = var_structure,
                     method = "ML",
                     data = data_df)
```

```r
partial_model2 = gls(Concentration ~ Biomass + fTreatment + fNutrient + 
                                     Biomass:fTreatment + 
                                     fTreatment:fNutrient,
                     weights = var_structure,
                     method = "ML",
                     data = data_df)
```

```r
partial_model3 = gls(Concentration ~ Biomass + fTreatment + fNutrient + 
                                     Biomass:fTreatment + 
                                     Biomass:fNutrient,
                     weights = var_structure,
                     method = "ML",
                     data = data_df)
```

```r
anova(full_model, partial_model1)
```

```r
anova(full_model, partial_model2)
```

```r
anova(full_model, partial_model3)
```

The above results indicate we can drop `Biomass:fTreatment`. We then proceed to dropping each of the two interaction terms and then based on these results the main term for `Biomass`. We are checking to see if AIC difference is significant. If there is an increase and the increase is significant, we are worse-off making the change

```r
final_model = gls(Concentration ~ fTreatment * fNutrient,
                  weights = var_structure,
                  method = "REML",
                  data = data_df)
```

```r
plot(final_model, col = 1)
```

```r
summary(final_model)
```

```r
boxplot(predict(final_model) ~ fTreatment * fNutrient, data = data_df)
```

# 5. Final recommendations

- Start with simple linear regression with as many predictors as possible. Check for deviations from the homogeneity of variance assumption by plotting residuals vs fitted values and residuals vs each predictor
- If violations are detected specify the variance structure from findings in step 1. Repeat this step till there is no observable pattern in residuals 
- After arriivng upon the variance structure that fixes the residual pattern conduct model selection
