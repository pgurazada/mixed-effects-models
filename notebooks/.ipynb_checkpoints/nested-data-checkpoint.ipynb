# 1. Objective

Explore mixed-effects models for data that follow a natural hierarchy. Datasets of this type violate the assumption of independence of observations in linear regression models

# 2. Hierarchical data

A basic problem we wish to counter here is that observations within a subgroup resemble each other more closely than those from another subgroup. There are two ways to tackle this problem - two stage models and hierarchical models

## 2.1. Two stage model

```r
data_df = read.delim("data/RIKZ.txt")
```

```r
glimpse(data_df)
```

In the first stage, we estimate the regression of `Richness` on `NAP` for each of the beaches.

```r
number_of_beaches = length(unique(data_df$Beach))
```

```r
betas_by_beach = vector(length = number_of_beaches)

for (beach_index in 1:number_of_beaches) {
    model_for_beach_i = summary(lm(Richness ~ NAP, 
                                   data = filter(data_df, Beach == beach_index))) 
    betas_by_beach[beach_index] = model_for_beach_i$coefficients[2, 1]
}
```

```r
betas_by_beach
```

In the second-stage, we estimate a regression model with the betas as the outcome.

```r
unique(data_df$Exposure)
```

```r
data_df %>%
    group_by(Beach) %>%
    summarize(mean_exposure = mean(Exposure)) %>%
    mutate(beta = betas_by_beach) ->
    beach_df
```

```r
anova(lm(beta ~ mean_exposure, data = beach_df))
```

The two-stage model is clunky and uses data summary for the second stage.

## 2.2. Hierarchical model

### 2.2.1. Random intercepts

We allow the intercept of the equation between Richness and NAP to vary by beach. Rather than introducing a set of regressoin coefficients like we did in the two-stage model, we model the variance between beaches as an additional parameter to be estimated.

```r
random_intercept_model1 = lme(Richness ~ NAP,
                              random = ~ 1 | factor(Beach),
                              data = data_df)
```

```r
plot(random_intercept_model1, col = 1)
```

```r
summary(random_intercept_model1)
```

```r
cat("Intra-class correlation (ICC) =", 2.944065^2/(2.944065^2 +3.05977^2))
```

### 2.2.2. Random intercepts and slopes

```r
random_int_and_slope_model1 = lme(Richness ~ NAP,
                                  random =  ~ 1 + NAP | factor(Beach),
                                  data = data_df)
```

```r
plot(random_int_and_slope_model1, col = 1)
```

### 2.2.3. Random effects

```r
random_effects_model1 = lme(Richness ~ 1,
                            random = ~ 1 | factor(Beach),
                            data = data_df)
```

```r
plot(random_effects_model1, col = 1)
```

### 2.2.3. Model selection

Start with a maximal fixed-effects specification and firm up the random-effects structure using AIC. Then, do a step-selection on the fixed effects keeping the random-effects components fixed.

```r
base_model1 = gls(Richness ~ 1 + NAP * factor(Exposure),
                  method = "REML",
                  data = data_df)
```

```r
random_intercepts_model2 = lme(Richness ~ 1 + NAP * factor(Exposure),
                               random = ~ 1 | factor(Beach),
                               data = data_df,
                               method = "REML")
```

```r
random_int_slope_model2 = lme(Richness ~ 1 + NAP * factor(Exposure),
                              random = ~ 1 + NAP | factor(Beach),
                              data = data_df,
                              method = "REML",
                              control = lmeControl(niterEM=5200, msMaxIter=5200))
```

```r
AIC(base_model1, random_intercepts_model2, random_int_slope_model2)
```

# 3. Full-length example

```r
data_df = read.delim("data/Owls.txt")
```

```r
glimpse(data_df)
```

```r
lm1 = lm(NegPerChick ~ SexParent  * FoodTreatment + SexParent * ArrivalTime,
         data = data_df)
```

```r
plot(fitted(lm1), residuals(lm1))
```

The plot again shows signs of heteroscedascity but something that cannot be fixed with an observation on the variance structure ascribed to a predictor. There is also evidence of clipping at a lower bound. 

```r
data_df %>% 
    mutate(log_neg= log10(NegPerChick + 1)) -> # Actually changing the outcome this way is a bad way of accounting for things
    data_df
```

```r
lm2 = lm(log_neg ~ SexParent  * FoodTreatment + SexParent * ArrivalTime,
         data = data_df)
```

```r
boxplot(rstandard(lm2) ~ Nest, 
        data = data_df,
        axes = FALSE,
        ylim = c(-3, 3))
abline(0, 0)
```

The residual pattern does not change to anything better because of the transformation

```r
model_formula = formula(log_neg ~ SexParent * FoodTreatment + SexParent * ArrivalTime)
```

```r
gls_model1 = gls(model_formula, 
                 data = data_df,
                 method = "REML")
```

```r
random_intercepts_model3 = lme(model_formula,
                               random = ~ 1 | Nest,
                               data = data_df,
                               method = "REML")
```

```r
random_int_slopes_model3 = lme(model_formula,
                               random = ~ 1 + ArrivalTime|Nest,
                               data = data_df,
                               method = "REML")
```

```r
AIC(gls_model1, random_intercepts_model3, random_int_slopes_model3)
```

```r
plot(random_int_slopes_model3, col = 1)
```

```r
summary(random_int_slopes_model3)
```

We can now take out the interaction terms one by one

```r
full_model = lme(model_formula,
                 random = ~ 1 + ArrivalTime | Nest,
                 method = "ML",
                 data = data_df)

ranef_model1 = update(full_model, . ~ . -SexParent:FoodTreatment)

ranef_model2 = update(full_model, . ~ . -SexParent:ArrivalTime)
```

```r
anova(full_model, ranef_model1)
```

```r
anova(full_model, ranef_model2)
```

```r
final_model1 = lme(log_neg ~ FoodTreatment + ArrivalTime,
                   random = ~ 1 + ArrivalTime | Nest,
                   method = "REML",
                   data = data_df)
```

```r
summary(final_model1)
```

```r
plot(final_model1, col = 1)
```

The plot is a bit suspicious. Variance still seems heterogeneous

```r
final_model2 = mgcv::gamm(log_neg ~ FoodTreatment + s(ArrivalTime),
                          random = list(Nest =~ 1),
                          data = data_df)
```

```r
summary(final_model2$gam)
```

```r
summary(final_model2$lme)
```

```r
plot(final_model2$gam)
```

```r
plot(final_model2$lme, col = 1)
```

Overall the result is still unsatisfying
