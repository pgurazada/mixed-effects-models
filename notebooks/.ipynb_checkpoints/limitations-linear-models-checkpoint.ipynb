# 1. Objective

Present examples where vanilla linear models are insufficient

# 2. Linear model assumption violations

Linear model assumptions are often invalid in real-life examples. There might be deviations from normality (screwing up standard error estimates), independence or homogeneity of variance. Further, models might be mis-specified when the relationship is assuemd to be linear.

## 2.1. Example 1 - Species density and nutrient generation

```r
data_df = read.delim("data/Nereis.txt")
```

```r
glimpse(data_df)
```

```r
dotchart(data_df$concentration,
         groups = factor(data_df$nutrient),
         pch = data_df$nutrient,
         ylab = "Order of observatios",
         xlab = "Concentration")
```

The above plot indicates that the assumption of homogeneity of variance across different groups in the data is violated

```r
pairs(data_df)
```

```r
boxplot(concentration ~ factor(nutrient),
        data = data_df,
        varwidth = TRUE,
        xlab = "Nutrient",
        ylab = "Concentration")
```

The box plot confirms difference in means, spread is a bit difficult to make out.

```r
nereiss_lm1 = lm(concentration ~ factor(biomass) * factor(nutrient),
                 data = data_df)
```

```r
drop1(nereiss_lm1, test = "F")
```

```r
summary(nereiss_lm1)
```

## 2.2. Example 2 - Biomass of wedge clams

```r
data_df = read.delim("data/Clams.txt")
```

```r
glimpse(data_df)
```

```r
data_df %>%
    mutate(ln_afd = log(AFD),
           ln_length = log(LENGTH),
           month = factor(MONTH)) -> 
    data_df
```

```r
coplot(ln_afd ~ ln_length | month, 
       data = data_df)
```

The above plot shows that the relationship between `ln_afd` and `ln_length` is different across the different months during which the observations were made.

```r
clam_lm1 = lm(ln_afd ~ ln_length * month,
              data = data_df)
```

```r
drop1(clam_lm1, test = "F")
```

```r
plot(clam_lm1, add.smooth = FALSE, which = 1)
hist(resid(clam_lm1), xlab = "Residuals", main = "")
plot(data_df$ln_length, 
     resid(clam_lm1), 
     xlab = "log(length)", 
     ylab = "residuals")
plot(data_df$month, 
     resid(clam_lm1), 
     xlab = "Month", 
     ylab= "residuals")
par(op)
```

```r
bartlett.test(resid(clam_lm1), data_df$month)
```

```r
summary(clam_lm1)
```

## 2.4. Pelagic bioluminescence

```r
data_df = read.delim("data/ISIT.txt")
```

```r
glimpse(data_df)
```

```r
xyplot(Sources ~ SampleDepth | factor(Station),
       data = data_df,
       xlab = "Sample Depth", 
       ylab = "Sources", 
       strip = function(bg = 'white', ...) strip.default(bg = 'white', ...), 
       panel = function(x, y) { 
                   panel.grid(h = -1, v = 2)
                   I1 <- order(x)
                   llines(x[I1], y[I1], col = 1)})
```

The non-linearity is apparent in the plots
